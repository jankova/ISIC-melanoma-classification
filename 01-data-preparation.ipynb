```python
# Extracting META features

import numpy as np
import config
import pandas as pd

train = pd.read_csv(config.TRAIN_PATH + 'train.csv')
train
```

```python
train_aug = pd.concat([train.drop('anatom_site_general_challenge', axis = 1), pd.get_dummies(train['anatom_site_general_challenge'], 
                                                    dtype=np.uint8, 
                                                    prefix='site')
                      ], axis = 1)
```

```python
train_aug
```

```python
meta_features = ['sex', 'age_approx'] + [col for col in train_aug.columns if 'site_' in col]
meta_features
```
